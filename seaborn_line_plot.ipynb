```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns # pip install seaborn
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/museum_visitors.csv", index_col = 0)
df.head()
```

```python
df.shape
```

```python
sns.lineplot(data = df)
plt.show()
```

```python
plt.figure(figsize = (16, 6))
plt.xticks(rotation = 90)
plt.title("Muzeylarga tashriflar soni")
plt.xlabel("Sana")
plt.ylabel("Odamlar soni")
plt.grid(color = 'k', alpha = 0.15)
sns.lineplot(data = df['Avila Adobe'])
plt.show()
```
