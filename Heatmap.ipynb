```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/museum_visitors.csv",index_col = 0, parse_dates = [0], usecols=[0, 1])
df.head()
```

```python
df['month'] = [i.month for i in df.index]
df['year'] = [i.year for i in df.index]
df.head()
```

```python
df2 = df.groupby(['month', 'year']).max()
df2
```

```python
df2.unstack(level = 1)
```

```python
plt.figure(figsize = (10,4))
sns.heatmap(data = df2.unstack(level=0),cmap = "Blues", linewidth = 1 ,square=True)
xticks_labels =["January", "February", "March", "April", "May",
                "June", "July", "August", "September", "October", 'November', "December"]

plt.xticks(np.arange(12) + .5, labels = xticks_labels)
plt.show()
```
